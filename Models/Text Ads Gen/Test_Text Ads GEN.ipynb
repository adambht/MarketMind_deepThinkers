```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch
from google.colab import drive
from huggingface_hub import login

# Replace 'your_token_here' with your actual Hugging Face token
login(token="")

# 1. Mount Google Drive
drive.mount("/content/drive")

# 2. Define the path to your saved model
model_path = "/content/drive/MyDrive/ad_models/models/mistral-7b-instruct-v0.3-finetuned"

# 3. Define the quantization config (same as used during training)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# 4. Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_path)

# 5. Load the quantized model from your Drive
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    quantization_config=bnb_config,
    device_map="auto"  # Automatically chooses the best device (GPU/CPU)
)

print("✅ Model and tokenizer loaded successfully.")

```

```python
# Function to get dynamic input for product name and description
def generate_ad_copy():
    # Get the product name and description from user input
    product_name = input("Enter product name: ")
    product_description = input("Enter product description: ")

    # Format the prompt
    prompt = f"Product: {product_name}\nDescription: {product_description}\nAd Copy:"

    # Tokenize the prompt
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # Generate a response using the model
    outputs = model.generate(**inputs, max_new_tokens=100, temperature=0.7)

    # Decode the generated text
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Print the result in a clear and well-presented format
    print("\n============================== Generated Ad ==============================")
    print(f"Product: {product_name}\nDescription: {product_description}")
    print("\nGenerated Ad :")
    print(generated_text)
    print("==============================================================================\n")

# Call the function to generate the ad copy
generate_ad_copy()
```
